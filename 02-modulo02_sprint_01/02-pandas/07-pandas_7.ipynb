### 1. Cread una columna nueva y una funci√≥n que nos d√© el mes en el que ocurri√≥ el ataque. 

Tened en cuenta que no todas las filas tienen la misma estructura 
y que puede que no haya la informaci√≥n de mes. En ese caso devolved un nulo (NaN).
üí° Pista üí°Pod√©is usar regex.

```python
date_unique = df["date"].unique()
date_unique
```

```python
def fecha(x):
    try:
        return re.findall('([A-Z]{1}[a-z]{2})\-',x)[0]
    except:
        return np.nan
```

```python
df['mes'] = df['date'].apply(fecha)
```

```python
df.head(2)
```

### 2. Vamos con la columna fatal_(y/n). 
Lo primero que ten√©is que hacer es evaluar los valores √∫nicos que hay. Esperar√≠amos tener solo dos valores: y y n, pero tristemente no ser√° as√≠ üò≠. Cread una columna nueva y una funci√≥n que devuelva √∫nicamente y o n y que devuelva un nulo (NaN) si no se encuentra el valor.


```python
df["fatal_(y/n)"].unique()
```

```python
def yn(x):
    if x == "N":
        return "N"
    elif x == "Y":
        return "Y"
    elif x == " N":
        return "N"
    elif x == "N ":
        return "N"
    elif x == "y":
        return "Y"
    else:
        return np.nan
```

```python
df["fatal"] = df["fatal_(y/n)"].apply(yn)
df.head(2)
```

```python
df["fatal"].unique()
```

### 3. Seguid la misma l√≥gica que en el ejercicio anterior pero con la columna sex. 
Pero en este caso vamos a incluir un retito m√°s: En este caso querremos tener dos valores √∫nicos F y M. De nuevo en la columna hay otros car√°cteres raros como 'M ', 'lli', 'N', '.'. En esos casos, queremos que se reemplacen por M o F de forma aleatoria (tendr√©is que usar la librer√≠a random). En caso de que sea nulo, que la funci√≥n devuelva nulo (np.nan).


```python
df["sex"].unique()
```

```python
def reemplazar_mf(x):
    if x == "F":
        return "F"
    elif x == "M":
        return "M"
    else:
        return np.nan
```

```python
df["sex_fm"] = df["sex"].apply(reemplazar_mf)
df.head(5)
```

```python
df["sex_fm"].unique()
```
