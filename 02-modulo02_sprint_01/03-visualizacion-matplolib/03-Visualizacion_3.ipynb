3. ¿Cuántos **ataques** hubo por **año**? Seleccionad solo desde el año **1500**.

```python
df_year_1500 = df[df["year"] >= 1500.0]
df_year_1500.head(2)
```

```python
plt.figure(figsize = (40,15))

fig = sns.countplot(x = "year", data = df_year_1500, palette = "husl")

fig.set(xlabel = "Año", ylabel = "Ataques")

fig.set_title("Ataques por año")

plt.xticks(rotation = 90)

plt.show()
```

### **Análisis**: 
Como podemos observar al obtener estos resultados, hubo un gran número de ataques en los años sesenta en comparación con siglos anteriores, una posterior bajada, y una resubida sin precedentes a partir del año 2000.

4. Seleccionad solo los registros de **USA, AUSTRALIA y SOUTH AFRICA.**

    Muestra el número de ataques por **país** y año desde el año **1900**.



```python
df_year_1900 = df[df["year"] >= 1900.0]
df_year_1900.head(2)
```

```python
df_aus_usa_sa = df_year_1900[df_year_1900['country'].isin(['USA','AUSTRALIA','SOUTH AFRICA'])]
df_aus_usa_sa.head(2)
```

```python
plt.figure(figsize = (10, 9.5))

fig = sns.countplot(x = "country", hue = "year", data = df_aus_usa_sa, palette = "husl")

fig.set(xlabel = "País", ylabel = "Ataques")

fig.set_title("Ataques por año")

fig.legend(bbox_to_anchor =(1,1), ncol = 3)
```

### **Análisis**

Aquí podemos comprobar que el país en el que más ataques se producen y también el país en el que más han crecido es USA. Australia experimenta una pequeña crecida también y Sudáfrica se mantiene uniforme.

```python
plt.figure(figsize =(10,6))

fig = sns.countplot(x = "country", data = df_aus_usa_sa, palette = "husl")

fig.set(xlabel = "País", ylabel = "Año")

fig.set_title("Ataques por año")

plt.xticks(rotation = 0)

plt.show()
```

### **Análisis**: 
Al parecer, el país en el que más ataques de tiburones hay es Estados Unidos, casi el doble de su siguiente, Australia. Sin embargo, esto podría ser debido a la cantidad de población de un país respecto a otro.

5. Realizad un boxplot.

    Donde se muestre los principales estadísticos del número de **ataques por país**.

```python
df_ataques_pais = df_aus_usa_sa.groupby(["country","year"])["case_number"].count().reset_index()
df_ataques_pais
```

```python
plt.figure(figsize = (8,6)) 

fig = sns.boxplot(x = "country", y = "case_number", data = df_ataques_pais, palette = "husl")

fig.set(xlabel = "País", ylabel = "Número de ataques")

plt.show()
```

### **Análisis**: 
Como podemos observar al obtener estos resultados, el mayor número de ataques se produce en USA, como decimos, seguramente por motivo de su gran población. En Sudáfrica las cifras ni suben ni bajan, y en Australia hay un crecimiento de los ataques que se aprecia en la gráfica anterior.
